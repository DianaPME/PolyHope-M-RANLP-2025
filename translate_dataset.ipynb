```python
import pandas as pd
import torch
from transformers import MarianMTModel, MarianTokenizer

# Load model for Spanish â†’ Urdu
model_name = 'Helsinki-NLP/opus-mt-en-ur'
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)

# Move model to GPU if available
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = model.to(device)

# Load your file
df = pd.read_csv('en_train.csv')

# Translation function
def translate_es_to_ur(text):
    inputs = tokenizer.prepare_seq2seq_batch([text], return_tensors="pt").to(device)
    translated = model.generate(**inputs)
    return tokenizer.decode(translated[0], skip_special_tokens=True)

# Apply translation
df['text_ur'] = df['text'].apply(translate_es_to_ur)

# Save output
df.to_csv('en_to_ur_train.csv', index=False)
```

```python
import pandas as pd
import torch
from transformers import MarianMTModel, MarianTokenizer

# Load Spanish to English model
model_name = 'Helsinki-NLP/opus-mt-es-en'
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)

# Move model to GPU if available
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = model.to(device)

# Load your file
df = pd.read_csv('/content/es_train.csv')

# Translation function
def translate(text):
    inputs = tokenizer.prepare_seq2seq_batch([text], return_tensors="pt").to(device)
    translated = model.generate(**inputs)
    return tokenizer.decode(translated[0], skip_special_tokens=True)

# Translate all rows
df['text_en'] = df['text'].apply(translate)

# Save the translated file
df.to_csv('es_to_en_train.csv', index=False)
```
